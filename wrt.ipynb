```python
import mediapipe as mp
import cv2
import numpy
```

```python
han=mp.solutions.hands
```

```python
cap=cv2.VideoCapture(0)
cap.set(3,1920)
cap.set(4,1080)
with han.Hands(
     max_num_hands=2
) as hn:
 pts=[]
 
 while cap.isOpened():
    bl,img=cap.read()
    img.flags.writeable=False
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    h,w=img.shape[:2]
    annote=hn.process(img)
    img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    img.flags.writeable=True
    # print(annote)
    if annote.multi_hand_landmarks:
        for hand_landmarks in annote.multi_hand_landmarks:
            x=int(hand_landmarks.landmark[han.HandLandmark.INDEX_FINGER_TIP].x * w)
            y=int(hand_landmarks.landmark[han.HandLandmark.INDEX_FINGER_TIP].y * h)
            y2=int(hand_landmarks.landmark[han.HandLandmark.MIDDLE_FINGER_TIP].y * h)
            x2=int(hand_landmarks.landmark[han.HandLandmark.MIDDLE_FINGER_TIP].x * w)
            if(y2>y+20):
             pts.append([x,y])
            else:
                if(x-x2<40):
                    for a in range (0,11):
                        for b in range (0,11):
                            if([x+a,y+b] in pts):
                                print(x,y)
                                pts.remove([x+a,y+b])
                            if([x-a,y+b] in pts):
                                print(x,y)
                                pts.remove([x-a,y+b])
                            if([x+a,y-b] in pts):
                                print(x,y)
                                pts.remove([x+a,y-b])
                            if([x-a,y-b] in pts):
                                print(x,y)
                                pts.remove([x-a,y-b])
    # if not annote.multi_hand_landmarks:
    #   continue
    # for hndl in annote.multi_hand_landmarks:
    #     f_x=int(hndl.landmark[han.HandLandmark.INDEX_FINGER_TIP].x * w)
    #     f_y=int(hndl.landmark[han.HandLandmark.INDEX_FINGER_TIP].y * h)
    #     print(f_x,f_y)
    #     pts.append([f_x,f_y])
    for pt in pts:
        cv2.circle(img,(pt[0],pt[1]),10,(255,0,0),cv2.FILLED)
    img=cv2.flip(img,1)
    cv2.imshow('ld',img)
    if cv2.waitKey(1)&0XFF==ord('a'):
        break
 cv2.destroyAllWindows()
```
